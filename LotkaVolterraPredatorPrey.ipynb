```python
import matplotlib.pyplot as plt
import numpy as np

def rk4(r, t, h):                  
        """ Метод Рунги-Кути """
        k1 = h*f(r, t)
        k2 = h*f(r+0.5*k1, t+0.5*h)
        k3 = h*f(r+0.5*k2, t+0.5*h)
        k4 = h*f(r+k3, t+h)
        return (k1 + 2*k2 + 2*k3 + k4)/6

def f(r, t):
        #g = 0.3 ### за відсутності хижаків чисельність жертв зростає експоненціально з коефіцієнтом g
        #a = 0.02 ### параметр, що описує зменшення чисельності жертв внаслідок їх зустрічей з хижа¬ками a 
        #s = 0.3 ### за відсутності здобичі хижаки вимирали б експоненціально з коефіцієнтом  s 
        #b = 0.004 ### параметр, що описує зростання чисельності хижа¬ків завдяки їх зустрічам з жертвами b
        g = 0.2 
        a = 0.06 
        s = 0.4 
        b = 0.003 
        x, y = r[0], r[1]
        fyd = b*x*y - s*y
        fxd = x*g - a*x*y
        return np.array([fxd, fyd], float)

h=0.025 ### 100 / 4000      

tpoints = np.arange(0, 100, h)         
xpoints, ypoints  = [], []
r = np.array([100, 40], float)

for t in tpoints:
        xpoints.append(r[0])          
        ypoints.append(r[1])          
        r += rk4(r, t, h)

plt.plot(tpoints, xpoints)
plt.plot(tpoints, ypoints)
plt.xlabel("Час")
plt.ylabel("Популяція")
plt.legend(('Жертви', 'Хижаки'))
plt.title("Рівняння Лотки—Вольтерри")
plt.savefig("Lotka_Volterra.png")
plt.show()

plt.plot(xpoints,ypoints)
plt.xlabel('Чисельність жертв')
plt.ylabel('Чисельність хижаків')
plt.title('Фазовий простір')
plt.show()
```
